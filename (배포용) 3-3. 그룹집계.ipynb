```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
# 실습 데이터 적재
data = pd.read_excel('data/인구수예제.xlsx')
data.head()
```

```python
# 실습 1. 연도별 총인구수 구하기
data.groupby('연도')['총인구'].sum()
```

```python
# 실습 2. 연도별 전체의 남자인구, 여자인구, 총인구 수 구하기
data.groupby('연도').sum()
```

```python
data.groupby('연도')[['남자인구','여자인구','총인구']].sum()
```

```python
# 실습 3. 자치구별로 평균 총인구수 구하기
data.groupby('자치구')['총인구'].mean()
```

```python
# 실습 4.도시/자치구별 평균 총인구수 구하기
data.groupby(['도시','자치구'])['총인구'].mean()
```

```python
# 실습 5. 도시/자치구별 평균 남자인구와 여자인구수 구하기
data.groupby(['도시','자치구'])[['남자인구','여자인구']].mean()
```

```python
# 실습 6. 도시별, 연도별로 총인구수 출력
data.groupby(['도시','연도'])['총인구'].sum().unstack(0)#멋있어!!
```

```python
data.pivot_table(index = '연도',
                columns = '도시',
                values = '총인구',
                aggfunc = ['sum','mean'])
```

```python
# 실습 7. 도시/자치구 별로 남자인구의 평균을 구한 후, 남자 인구가 가장 많은 도시 및 자치구 Top 5 찾기
data.pivot_table(index=['도시','자치구'],
                values = '남자인구',
                aggfunc = 'mean').sort_values(by= '남자인구',ascending=False)[:5]
```

```python
data.pivot_table(index=['도시','자치구'],
                values = '남자인구',
                aggfunc = 'mean').nlargest(5,'남자인구')
```

```python
# 실습 8. 여자가 남자보다 많은 도시/자치구 상위 3개 찾기 
ex = data.pivot_table(index = ['도시','자치구'],
                values = ['남자인구','여자인구'],
                aggfunc = 'mean')
```

```python
ex['남녀차이'] = ex.여자인구 -ex.남자인구
ex
```

```python
ex.nlargest(3,'남녀차이')
```
